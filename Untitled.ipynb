# Armstrong 

```python
def armstrong():
    num = int(input('Enter The Number = '))
    digits_of_number = sum(int(digit) ** len(str(num)) for digit in str(num))

    if num == digits_of_number:
        print("This is Armstrong Number")
    else:
        print("This is Not Armsrong Number")
    
```

# Perfect Number

```python
def perfect_number():
    num = int(input('Enter The Number'))
    sum = 0
    for i in range(1, num):
        if num%i==0:
            sum+i
    if num==sum:
        print('this is perfect number')
    else:
        print('this is not perfect number')
```

```python
perfect_number()
```
